```python
cursor.execute("""
    CREATE TABLE customer_data (
        customer_id NVARCHAR(50),
        customer_unique_id NVARCHAR(50),
        customer_zip_code_prefix INT,
        customer_city NVARCHAR(50),
        customer_state NVARCHAR(10)
    )
""")
```

```python
for row in customer_data.itertuples():
    cursor.execute(''' 
    INSERT INTO [Data Warehouse].dbo.customer_data (customer_id, customer_unique_id, customer_zip_code_prefix, customer_city, customer_state)
    VALUES (?, ?, ?, ?, ?)
    ''', 
    (row.customer_id, row.customer_unique_id, row.customer_zip_code_prefix, row.customer_city, row.customer_state)
    )

conn.commit()
```

```python
cursor.execute("""
    CREATE TABLE order_payment_data (
        order_id VARBINARY(MAX),
        payment_sequential int,
        payment_type NVARCHAR(50),
        payment_installments INT,
        payment_value float
        
    )
""")
```

```python
for row in order_payment_data.itertuples():
    cursor.execute(''' 
    INSERT INTO [Data Warehouse].dbo.order_payment_data (order_id, payment_sequential, payment_type, payment_installments, payment_value)
    VALUES (
        CONVERT(VARBINARY(MAX), ?),  -- order_id
        ?,                           -- payment_sequential
        CONVERT(NVARCHAR(MAX), ?),   -- payment_type (assuming it's NVARCHAR)
        ?,                           -- payment_installments
        ?                            -- payment_value
    )
    ''', 
    row.order_id,
    row.payment_sequential,
    row.payment_type,
    row.payment_installments,
    row.payment_value
    )
    
conn.commit()
```
